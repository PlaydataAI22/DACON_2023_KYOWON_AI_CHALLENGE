```python
img_paths = [r"D:\im\project\parseq\trainset\test", r"D:\im\project\parseq\trainset\train"]
```

```python
target_height = 32
target_width = 100
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np




for path in img_paths:
    for fname in os.listdir(path):
        i = cv2.imread(os.path.join(path, fname), cv2.IMREAD_GRAYSCALE)
        height, width = i.shape
        print(i.shape)
        y_per = target_height/height
        if y_per * width > target_width:
            y_per = target_width/width
            i = cv2.resize(i, (0, 0), fx=y_per, fy=y_per, interpolation=cv2.INTER_AREA)
        else:
            i = cv2.resize(i, (0, 0), fx=y_per, fy=y_per, interpolation=cv2.INTER_AREA)
        i_height, i_width = i.shape
        print(i.shape)
        i = np.pad(i, ((0, 32-i_height), (0, 100-i_width)), mode="constant", constant_values=np.median(i))
        plt.imshow(i, cmap="gray")
        plt.show()
```
